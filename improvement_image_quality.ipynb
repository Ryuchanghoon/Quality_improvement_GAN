```python
import torch
import os
import random
from PIL import Image
import torchvision
import torchvision.transforms as transforms
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
transform = transforms.Compose([transforms.ToTensor()])

dataset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
```

```python
# 폴더 생성.
if not os.path.exists('saved_images'):
    os.makedirs('saved_images')
```

```python
for i in range(1000):
    image_tensor, label = random.choice(dataset)
    image = transforms.ToPILImage()(image_tensor)
    image_path = os.path.join('saved_images', f'image_{i}.png')
    image.save(image_path)
    print(f"Image {i} saved to {image_path}")
```

```python
source_folder = '/content/saved_images'
color_folder = 'CIFAR10_color_images'
gray_folder = 'CIFAR10_gray_images'
```

```python
if not os.path.exists(color_folder):
    os.makedirs(color_folder)

if not os.path.exists(gray_folder):
    os.makedirs(gray_folder)
```

```python
# 소스 이미지 불러오기
image_files = [f for f in os.listdir(source_folder) if os.path.isfile(os.path.join(source_folder, f))]
image_files.sort()  # 정렬
```

```python
for i in range(500):
    image_path = os.path.join(source_folder, image_files[i])
    target_path = os.path.join(color_folder, image_files[i])
    os.rename(image_path, target_path)
```

```python
for i in range(500, 1000):
    image_path = os.path.join(source_folder, image_files[i])
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # 흑백으로 로드
    target_path = os.path.join(gray_folder, image_files[i])
    cv2.imwrite(target_path, image)  # 흑백 이미지 저장
```

```python
### 해상도 개선 EDSR 모델 로드

sr = cv2.dnn_superres.DnnSuperResImpl_create()
sr.readModel('/content/drive/MyDrive/GAN 실습/EDSR_x3.pb')
sr.setModel('edsr', 3)
```

```python
color_folder = '/content/CIFAR10_color_images'
gray_folder = '/content/CIFAR10_gray_images'

improved_color_folder = 'imporved_color_images'
improved_gray_folder = 'improved_gray_images'


if not os.path.exists(improved_color_folder):
    os.makedirs(improved_color_folder)


if not os.path.exists(improved_gray_folder):
    os.makedirs(improved_gray_folder)

# 이미지 파일 로드
color_image_files = [f for f in os.listdir(color_folder) if os.path.isfile(os.path.join(color_folder, f))]
color_image_files.sort()


gray_image_files = [f for f in os.listdir(gray_folder) if os.path.isfile(os.path.join(gray_folder, f))]
gray_image_files.sort()



for i in range(500):
    color_image_path = os.path.join(color_folder, color_image_files[i])
    color_img = cv2.imread(color_image_path)
    improved_img = sr.upsample(color_img)
    color_target_path = os.path.join(improved_color_folder, color_image_files[i])
    cv2.imwrite(color_target_path, improved_img)


for i in range(500):
    gray_image_path = os.path.join(gray_folder, gray_image_files[i])
    gray_img = cv2.imread(gray_image_path, cv2.IMREAD_GRAYSCALE)
    improved_img = sr.upsample(gray_img)
    gray_target_path = os.path.join(improved_gray_folder, gray_image_files[i])
    cv2.imwrite(gray_target_path, improved_img)
```
